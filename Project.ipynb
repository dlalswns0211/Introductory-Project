```python
# 파생지표 생성 예시
# 어시스트 대비 턴오버 비율
details["AST_TO_ratio"] = details["AST"] / details["TO"].replace(0, 1)

# 총 리바운드 비율
details["REB_share"] = details["REB"] / (details["REB"] + details["OPP_REB"])

# 슈팅 효율
details["TS_pct"] = details["PTS"] / (2 * (details["FGA"] + 0.44*details["FTA"]))
```

```python
import matplotlib.pyplot as plt

# 승리 vs 패배에 따른 필드골 성공률 비교
games.boxplot(column="FG_PCT_home", by="HOME_TEAM_WINS")
plt.title("Field Goal % by Home Team Wins")
plt.suptitle("")
plt.show()

# 팀별 승률
team_winrate = ranking.groupby("TEAM")["W_PCT"].mean().sort_values(ascending=False)
team_winrate.plot(kind="bar", title="Average Team Win Percentage")
plt.show()
```
